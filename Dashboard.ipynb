```python
import networkx
import matplotlib.pyplot as plt
import numpy

%matplotlib ipympl
```

```python
graph = networkx.Graph()
```

```python
graph.clear()

graph.add_nodes_from([1, 2, 3, 4])

graph.add_edge(1, 3)
graph.add_edge(2, 3)

graph.add_edge(1, 4)
graph.add_edge(2, 4)

graph.add_edge(2, 1)
```

```python
subplot = plt.subplot()
networkx.draw(graph, with_labels=True)

plt.show()
```
