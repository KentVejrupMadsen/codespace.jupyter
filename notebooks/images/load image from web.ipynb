```python
from requests \
    import get \
    as get_http_request

from PIL \
    import Image

from io \
    import BytesIO

from IPython.display import display
```

```python
url_to_image: str = '''https://upload.wikimedia.org/wikipedia/commons/3/3e/Storegade_182%2C_Esbjerg%2C_2.jpg'''

headers = {
    'User-Agent': 'jupyter-script/0.1(https://github.com/KentVejrupMadsen/codespace.jupyter; kent.vejrup.madsen@designermadsen.dk)'
}
```

```python
response = get_http_request(
    url=url_to_image, 
    headers=headers
)
```

```python
if response.ok:
    with Image.open(BytesIO(response.content)) as image:
        width, height = image.size

        resize_width: int = int(width/10)
        resize_height: int = int(height/10)

        image = image.resize((resize_width, resize_height))

        display(image)
```
