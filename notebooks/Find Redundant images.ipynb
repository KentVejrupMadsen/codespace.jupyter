```python
from os \
    import listdir

from os.path \
    import \
        join, \
        isfile, \
        basename

import hashlib
```

# Variables

## Constants

```python
verify: bool = False

sha_1: str = '1'
sha_256: str = '256'
sha_512: str = '512'

zero: int = 0
```

```python
def get_constant_zero() -> int:
    global zero
    return zero
```

## Dynamics

```python
path_to_location: str = 'D:\\Photos\\for datasets'
read_buffer_size: int = 1024

files_found: list = []
files_in_memory: list = []


hash_1_generator = None
hash_256_generator = None
hash_512_generator = None
```

```python
listed_file_names: list = listdir(
    path_to_location
)

for idx in range(
    len(
        listed_file_names
    )
):
    full_location: str = join(
        path_to_location, 
        listed_file_names[idx]
    )

    if isfile(full_location):
        files_found.append(
            full_location
        )

del listed_file_names
```

## SHA Hash functions

```python
def get_sha1_hash(
        location_to_file: str,
        buffered_file_chunk
) -> str | None:
    global \
        read_buffer_size, \
        hash_1_generator
    
    if buffered_file_chunk is None:
        with open(
            location_to_file, 
            'rb'
        ) as opened_file:
            instance_generator = hashlib.sha1(
                usedforsecurity=False
            )

            chunk = get_constant_zero()
            
            while chunk != b'':
                chunk = opened_file.read(read_buffer_size)
                instance_generator.update(chunk)
        
            opened_file.close()

            return str(
                instance_generator.hexdigest()
            ).upper()
    else:
        if hash_1_generator is None:
            hash_1_generator = hashlib.sha1(
                usedforsecurity=False
            )
        
        hash_1_generator.update(buffered_file_chunk)        

    return str(
        hash_1_generator.hexdigest()
    ).upper()
```

```python
def get_sha256_hash(
        location_to_file: str,
        buffered_file_chunk
) -> str | None:
    global \
        read_buffer_size, \
        hash_256_generator
    
    if buffered_file_chunk is None:
        with open(
            location_to_file, 
            'rb'
        ) as opened_file:
            instance_generator = hashlib.sha256(
                usedforsecurity=False
            )

            chunk = get_constant_zero()
            
            while chunk != b'':
                chunk = opened_file.read(read_buffer_size)
                instance_generator.update(chunk)
            
            opened_file.close()
            
            return str(
                instance_generator.hexdigest()
            ).upper()
    else:
        if hash_256_generator is None:
            hash_256_generator = hashlib.sha256(
                usedforsecurity=False
            )
        
        hash_256_generator.update(buffered_file_chunk)
        
    return str(
        hash_256_generator.hexdigest()
    ).upper()
    
```

```python
def get_sha512_hash(
        location_to_file: str,
        buffered_file_chunk
) -> str:
    global \
        read_buffer_size, \
        hash_512_generator

    if buffered_file_chunk is None:
        with open(
            location_to_file, 
            'rb'
        ) as opened_file:
            instance_generator = hashlib.sha512(
                usedforsecurity=False
            )

            chunk = get_constant_zero()
            
            while chunk != b'':
                chunk = opened_file.read(read_buffer_size)
                instance_generator.update(chunk)
            
            opened_file.close()
            
            return str(
                instance_generator.hexdigest()
            ).upper()
    else:
        if hash_512_generator is None:
            hash_512_generator = hashlib.sha512(
                usedforsecurity=False
            )

        hash_512_generator.update(buffered_file_chunk)
        
    return str(
        hash_512_generator.hexdigest()
    ).upper()
```

```python
def clear_buffers():
    global \
        hash_1_generator,   \
        hash_256_generator, \
        hash_512_generator
    
    del \
        hash_1_generator,  \
        hash_256_generator, \
        hash_512_generator

    hash_1_generator = None
    hash_256_generator = None
    hash_512_generator = None
```

```python
class Hashes:
    def __init__(
            self, 
            file_location: str
    ):
        global       \
            sha_1,   \
            sha_256, \
            sha_512

        self.sha: dict = \
        {
            sha_1: None,
            sha_256: None,
            sha_512: None
        }

        self.generate(file_location)

    def generate(self, location):
        global read_buffer_size
        
        sha_1 = None
        sha_256 = None
        sha_512 = None

        with open(location, 'rb') as open_file:
            chunk = 0

            while chunk != b'':
                chunk = open_file.read(read_buffer_size)

                sha_1 = get_sha1_hash(
                    location_to_file=location, 
                    buffered_file_chunk=chunk
                )
                
                sha_256 = get_sha256_hash(
                    location_to_file=location, 
                    buffered_file_chunk=chunk
                )

                sha_512 = get_sha512_hash(
                    location_to_file=location, 
                    buffered_file_chunk=chunk
                )

            open_file.close()
        
        clear_buffers()

        self.set_sha_1(sha_1)
        self.set_sha_256(sha_256)
        self.set_sha_512(sha_512)


    def get_sha_1(self) -> str:
        global sha_1
        return self.sha[sha_1]
    
    def get_sha_256(self) -> str:
        global sha_256
        return self.sha[sha_256]
    
    def get_sha_512(self) -> str:
        global sha_512
        return self.sha[sha_512]
    
    def set_sha_1(
            self, 
            value: str
    ) -> None:
        global sha_1
        self.sha[sha_1] = value

    def set_sha_256(
            self, 
            value: str
    ) -> None:
        global sha_256
        self.sha[sha_256] = value

    def set_sha_512(
            self, 
            value: str
    ) -> None:
        global sha_512
        self.sha[sha_512] = value
    
    def __repr__(self) -> str:
        global sha_256

        return repr(
            self.sha[sha_256]
        )
```

```python
def get_file_extension(file_path: str) -> str | None:

    tokens = file_path.split('.')
    
    last_token = tokens[int(len(tokens) - 1)]

    return last_token
```

```python
class FileLoader:
    def __init__(
            self, 
            location_to_file: str
        ):
        self.location: str = location_to_file
        self.filename: str = basename(location_to_file)

        self.extension: str | None = None
        
        self.hashes: Hashes = Hashes(
            location_to_file
        )

        self.set_extension(
            get_file_extension(
                self.get_file_name()
            )
        )
    
    def get_file_name(self) -> str:
        return self.filename
    
    def set_file_name(self, value: str) -> None:
        self.filename = value

    def get_file_location(self) -> str:
        return self.location
    
    def set_file_location(self, value) -> None:
        self.location = value

    def get_extension(self):
        return self.extension    
    
    def get_hashes(self):
        return self.hashes
    
    def set_hashes(
            self, 
            value
        ) -> None:
        self.hashes = value

    def set_extension(
            self, 
            value: str
        ) -> None:
        self.extension = value

    def __repr__(self):
        return repr((
            self.hashes,
            self.extension,
            self.filename
        ))
```

```python
for file \
    in files_found:
    files_in_memory.append(
        FileLoader(
                file
            )
        )
```

### Verification

```python
if verify == True:
    for loaded in files_in_memory:
        test_by = get_sha1_hash(
            loaded.get_file_location(), 
            None
        )
        
        if not (loaded.get_hashes().get_sha_1() == test_by):
            print('ERROR')
            
            print(
                'compare to: ', 
                loaded.get_hashes().get_sha_1()
            )
            
            print(
                'test: ', 
                test_by
            )
```

```python
def optimize_loaded_files():
    global \
        files_in_memory, \
        sha_256

    sorted(
        files_in_memory, 

        key=lambda fileloader: fileloader.hashes.sha[sha_256] and \
                               fileloader.extension and \
                               fileloader.filename
    )

optimize_loaded_files()
```
