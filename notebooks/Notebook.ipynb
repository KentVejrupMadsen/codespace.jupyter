```python
%matplotlib ipympl

import networkx as xGraph
from random import SystemRandom
```

```python
random_generator: SystemRandom = SystemRandom()

def get_random_generator() -> SystemRandom:
    global random_generator
    return random_generator
```

```python
last_index: int = 1

def get_current_index() -> int:
    global last_index
    return last_index

def next_index() -> None:
    global last_index
    last_index = last_index + 1
```

```python
network_graph = xGraph.Graph()

first_layer = []
second_layer = []
```

```python
for index in range(10):
    current_index = get_current_index()
    first_layer.append(current_index)

    next_index()
```

```python
for index in range(20):
    current_index = get_current_index()

    second_layer.append(
        current_index
    )

    next_index()
```

```python
selected_in_first_layer: int = 0


for selected_index in range(len(second_layer)):
    currently_selected_node_in_second_layer: int = second_layer[selected_index]

    if selected_index % 2 == 0:
        selected_in_first_layer = selected_in_first_layer + 1

    position_of_index_in_first_layer: int = selected_in_first_layer - 1

    selected_node_in_first_layer: int = first_layer[position_of_index_in_first_layer]

    network_graph.add_edge(
        currently_selected_node_in_second_layer, 
        selected_node_in_first_layer
    )
```

```python
# Chain the graph

size_of_first_layer: int = len(first_layer)

for i in range(size_of_first_layer):
    currently_selected: int = first_layer[i]
    
    next_node: int = i + 1

    if next_node >= size_of_first_layer:
        next_node: int = next_node % size_of_first_layer
    
    next_node_in_set = first_layer[next_node]
    
    network_graph.add_edge(
        currently_selected, 
        next_node_in_set
    )

    
```

```python
xGraph.draw(network_graph, with_labels = True)
```

```python
if not (network_graph is None):
    network_graph.clear()
```
