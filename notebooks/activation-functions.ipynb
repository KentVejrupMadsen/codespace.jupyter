```python
from matplotlib import pyplot as plot
from math import e as eulers_number
from math import pow
```

```python
class SigmoidActivation:
    def __init__(
        self, 
        x: float = 0.0
    ) -> None:
        self.x: float = x
        self.move_x: float = -1.0

        self.height: float = 1.0
        self.depth: float = 2.0

    def __del__(self) -> None:
        del                 \
            self.x,         \
            self.height,    \
            self.depth,     \
            self.move_x
    
    def calculate(self) -> float:
        power_of: float = pow(
            eulers_number, 
            -(self.get_input())
        )

        bottom: float = self.get_height() + power_of
        return (self.calculate_top()/bottom) + self.move_x
    
    def calculate_top(self) -> float:
        top: float = 1
        top = top * self.depth

        return top

    def set_input(
            self, 
            x: float 
        ) -> None:
        self.x = x
    
    def get_input(self) -> float:
        return self.x
    
    def get_height(self) -> float:
        return self.height
```

```python
def calculate_sigmoid_activation_range(range_x: list) -> list:
    result: list = []
    
    activation = SigmoidActivation()

    for x in range_x:
        activation.set_input(x)
        result.append(activation.calculate())

    return result
```

```python
fx: list = list(range(-20, 20))
fy: list = calculate_sigmoid_activation_range(fx)

plot.clf()
plot.plot(fx, fy)
plot.show()
```
