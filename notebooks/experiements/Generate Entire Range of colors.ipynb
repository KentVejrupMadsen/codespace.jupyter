```python
from HardenedSteel.objects \
    import CounterObject

from PIL import Image

from IPython.display import display
from os.path import isdir, join
from os import mkdir

from multiprocessing import Pool
```

```python
!rm /tmp/output -R
```

```python
red: CounterObject = CounterObject(
    start_value=0
)

green: CounterObject = CounterObject(
    start_value=0
)

blue: CounterObject = CounterObject(
    start_value=0
)

save_at_directory: str = '/tmp/output/'

processes: int = 8

color_range: list = list()
red_range: list = list()
green_range: list = list()
blue_range: list = list()

pool_size: int = 0
rows: list = list()


def get_red_counter() -> CounterObject:
    global red
    return red

def get_green_counter() -> CounterObject:
    global green
    return green

def get_blue_counter() -> CounterObject:
    global blue
    return blue

def get_red() -> int:
    return int(
        get_red_counter()
    )

def get_green() -> int:
    return int(
        get_green_counter()
    )

def get_blue() -> int:
    return int(
        get_blue_counter()
    )

def get_pixel_color() -> tuple:
    return(
        get_red(), 
        get_green(), 
        get_blue()
    )

def red_is_at_end() -> bool:
    return get_red() >= 256

def green_is_at_end() -> bool:
    return get_green() >= 256

def blue_is_at_end() -> bool:
    return get_blue() >= 256


def red_is_at_beginning() -> bool:
    return get_red() == 0

def green_is_at_beginning() -> bool:
    return get_green() == 0

def blue_is_at_beginning() -> bool:
    return get_blue() == 0
```

```python
while not (red_is_at_end()):
    red_range.append(
        get_red()
    )
    get_red_counter().increment()

while not (green_is_at_end()):
    green_range.append(
        get_green()
    )
    get_green_counter().increment()

while not (blue_is_at_end()):
    blue_range.append(
        get_blue()
    )
    get_blue_counter().increment()

for red in red_range:
    for green in green_range:
        for blue in blue_range:
            color_range.append(
                (
                    red, 
                    green, 
                    blue
                )
            )

if not isdir(save_at_directory):
    mkdir(save_at_directory)
```

```python
length_of_set: int = len(color_range)
pool_size = int(length_of_set/processes)

for row in range(processes):
    result_position: int = int(row * pool_size)
    rows.append(result_position)
```

```python
print(pool_size)
```

```python
def generate_image(
        index: int,
        color: tuple
) -> None:
    global save_at_directory
    save_filename: str = str(index) + '.jpg'
    
    full_path: str = join(
        save_at_directory, 
        save_filename
    )

    with Image.new(
        'RGB', 
        (512, 512), 
        color=color
    ) as image:
        image.save(
            full_path,
            bitmap_format='jpg',
            optimize=True
        )
```

```python
def create_image(
    row_index_start
) -> None:
    global pool_size, color_range, length_of_set

    final: int = row_index_start + pool_size

    for index in range(
        row_index_start, 
        final
    ):
        color = color_range[index]

        generate_image(
            index=index,
            color=color
        )

        if index == length_of_set:
            break
```

```python
with Pool(processes) as process_pool:
    process_pool.map(
        create_image, 
        rows
    )
```
