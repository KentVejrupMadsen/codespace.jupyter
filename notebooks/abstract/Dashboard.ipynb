```python
from PIL import Image
from IPython.display import display
from math import sin, cos, tan
```

```python
width: int = int(1920/3)
height: int = int(1080/3)

canvas_size: tuple = (width, height)
```

```python
with Image.new(
    'RGB', 
    canvas_size, 'white'
) as image:
    pixel_accessor = image.load()

    for x in range(width):
        for y in range(height):
            red, green, blue = pixel_accessor[x, y]            
            pixel_accessor[x, y] = (red, green, blue)
    
    display(image)
```
