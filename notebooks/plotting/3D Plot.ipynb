```python
from matplotlib import pyplot as plot
from mpl_toolkits import mplot3d
import numpy
from math import sin, cos

%matplotlib ipympl
```

```python
fig = plot.figure()
ax = plot.axes(
    projection='3d'
)

x: list = []
y: list = []
z: list = []

def generate_graph() -> None:
    global x, y, z

    counter: float = 0.0
    move_by: float = 0.1

    for n in range(0, 450):
        counter = counter + move_by

        x.append(
            generate_x(
                counter
            )
        )
        y.append(
            generate_y(
                counter
            )
        )
        z.append(
            generate_z(
                counter
            )
        )

def generate_x(value) -> float:
    return cos(float(value))


def generate_y(value) -> float:
    return sin(float(value))


def generate_z(value) -> float:
    return float(value)


generate_graph()
ax.plot(x, y, z)
```
