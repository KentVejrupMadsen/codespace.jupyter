```python
from IPython.display import display, Image
from PIL import Image as PImage
from os.path import isfile
from os import remove
```

```python
file_path: str = './original.jpg'
save_to: str = './new.jpg'
```

```python
def make_thumbnail(
        image: PImage.Image, 
        scale_factor_width: float = 7.0, 
        scale_factor_height: float = 7.0
):
    scaled_down_width: int = int(
        float(
                image.width / scale_factor_width
        ))
    scaled_down_height: int = int(
        float(
                image.height / scale_factor_height
        ))

    thumbnail: PImage.Image = image.resize(
        (
                scaled_down_width, 
                scaled_down_height
        ))

    return thumbnail
```

```python
with PImage.open(file_path) as original_file:
    display(make_thumbnail(original_file))
```

```python
with PImage.open(file_path) as loaded_in_memory:
    if isfile(save_to):
        remove(save_to)

    loaded_in_memory.save(save_to, quality=5, subsampling=0)


with PImage.open('new.jpg') as new_file:
    display(make_thumbnail(new_file))
```
